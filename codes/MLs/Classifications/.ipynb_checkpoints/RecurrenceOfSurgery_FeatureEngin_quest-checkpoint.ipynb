```python
import pandas as pd
```

```python
df_ROS = pd.read_csv('../../../datasets/RecurrenceOfSurgery.csv')
df_ROS
```

#### 전처리 과정

```python
df_ROS.columns
```

```python
df_ROS.info()
```

```python
df_ROS_data = df_ROS[['재발여부', '직업','연령','입원일자']]
df_ROS_data
```

```python
df_ROS_data = df_ROS_data.dropna()
```

```python
df_ROS_data.isnull().sum()
```

#### Scaling & Encoding
- split 하기 전에 실시하는게 유리

##### Encoding with One Hot Encoding
- 범주형 데이터를 연속형 숫자 데이터로 변환
- 범주형 항목만큼 벡터화

```python
df_ROS_data['직업'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oneHotEncoder = OneHotEncoder() #인스턴스화 
oneHotEncoder.fit(df_ROS_data[['직업']]) ## value값만 넣어줘야 함 [[]]을 통해 value만 넣어줌
# 해당 항목 학습한 것임.

## .value_counts() 항목만 넣어줘도 됨 / 몇개의 항목이 있는지 모를경우 전체의 값을 집어넣음
```

```python
column_name = oneHotEncoder.categories_  # 컬럼 name
```

```python
# oneHotEncoder.transform(df_TFD_extract_preprocess[['Pclass']]).toarray() #values # 실제 값 확인용 
encoded_data = oneHotEncoder.transform(df_ROS_data[['직업']]).toarray()
encoded_data
```

```python
df_ROS_data[['직업']][:3]
```

```python
encoder_array = oneHotEncoder.transform(df_ROS_data[['직업']]).toarray()
encoder_array
```

```python
## DataFrame에 담기 위한 작업 
df_encoded_data = pd.DataFrame(encoder_array, columns=oneHotEncoder.get_feature_names_out(['직업']))
df_encoded_data[:2]
```

```python
df_ROS_concat = pd.concat([df_ROS_data.reset_index(drop=True),df_encoded_data.reset_index(drop=True)], axis =1)
df_ROS_concat
```

```python
df_ROS_concat.columns
```

```python
target = df_ROS_concat['재발여부']
features = df_ROS_concat.drop(columns=['재발여부', '직업'])
```

#### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
minMaxScaler = MinMaxScaler() #인스턴스 화
```

```python
features = minMaxScaler.fit_transform(features)
features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=111)
features_train.shape, target_train.shape, features_test.shape, target_test.shape
```

#### 모델학습

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_params = {'min_samples_leaf' : [5, 7, 9],'max_depth' : [9, 11],'min_samples_split' : [5, 6, 7] }
```

#### 평가

```python
from sklearn.metrics import f1_score
```

```python
grid_search = GridSearchCV(model, param_grid=hyper_params, cv=3, verbose=1, scoring=f1_score)
```

```python
grid_search.fit(features_train,target_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_, grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
target_test_predict = best_model.predict(features_test)
target_test_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target_test, target_test_predict))
```
